```python
from ultralytics import YOLO
import os

model = YOLO('best.pt')  

input_folder = "./Data/Images"
output_folder = "./Data/Labels"

images = [os.path.join(input_folder, i) for i in os.listdir(input_folder)]

results = model(images)

for image, result in zip(images, results):
    
    boxes = result.boxes
    image_name = image.split('/')[-1][:-4]
    
    for annotation, class_id in zip(boxes.xywhn, boxes.cls):
        x_center, y_center, width, height = annotation.cpu().numpy()
        annotation_line = f"{int(class_id)} {x_center:.7f} {y_center:.7f} {width:.7f} {height:.7f}\n"

        with open(os.path.join(output_folder, f'{image_name}.txt'), 'a') as f:
            f.write(annotation_line)
```

```python
from ultralytics import YOLO
import cv2
import math 
# start webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

# model
model = YOLO("./best.pt")

input_folder = "./Data/Images"
output_folder = "./Data/Labels"

# object classes
classNames = ["ball"]
ind=0
while True:
    success, img = cap.read()
    results = model(img, stream=True, conf=0.5)
    img2 = img.copy()

    for r in results:
        boxes = r.boxes

        for annotation, class_id in zip(boxes.xywhn, boxes.cls):
            x_center, y_center, width, height = annotation.cpu().numpy()
            annotation_line = f"{int(class_id)} {x_center:.7f} {y_center:.7f} {width:.7f} {height:.7f}\n"
    
            with open(os.path.join(output_folder, f'{ind}.txt'), 'a') as f:
                f.write(annotation_line)
            
        for box in boxes:
            # bounding box
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2) # convert to int values

            # put box in cam
            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 3)

            # confidence
            confidence = math.ceil((box.conf[0]*100))/100
            print("Confidence --->",confidence)

            # class name
            cls = int(box.cls[0])
            print("Class name -->", classNames[cls])

            # object details
            org = [x1, y1]
            font = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 1
            color = (255, 0, 0)
            thickness = 2

            cv2.putText(img, classNames[cls], org, font, fontScale, color, thickness)

    cv2.imshow('Webcam', img)
    if r:
        cv2.imwrite(os.path.join(input_folder, f'{ind}.jpg'), img2)
        ind+=1
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
